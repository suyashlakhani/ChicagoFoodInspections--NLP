```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv("Food_Inspections.csv")
```

**1. Extracting regulation descriptions from each record corresponding to a failed inspection**

```python
df_new = df[df["Results"]=="Fail"]
```

```python
df_new.shape
```

```python
df_final = df_new[~df_new["Violations"].isnull()]
df_final.shape
```

```python
df_final.head()
```

```python
import re
reg_discrip =[]
df_final['Violations'] = df_final['Violations'].astype(str) #Converting to string
df_final['Violations'] = df_final['Violations'].apply(lambda x: x.split('|')) #Splitting the violations
for text in df_final["Violations"]:
    for i in text:
        matches = re.findall(r'\s+[A-Z\W]+-', i) #regex to find the regulation description from the violation column
    for match in matches:
        reg_discrip.append(match)
        
len(reg_discrip)
```

**2. Tokenizing each regulation description**

```python
import nltk as nltk
import nltk.corpus  
from nltk.text import Text
nltk.download('punkt')
token_count={}
for text in reg_discrip:
    tokens = nltk.tokenize.word_tokenize(text)
    for token in tokens:
        if token in token_count:
            token_count[token]+=1
        else:
            token_count[token]=1

    
```

```python
token_count
```

**3. Finding top-10 tokens (for the whole table)**

```python
import operator 
sorted_token_count = sorted(token_count.items(),key=operator.itemgetter(1),reverse=True)

sorted_token_count = pd.DataFrame(sorted_token_count)
```

```python
sorted_token_count.head(10)
```

```python
sorted_token_count = sorted_token_count.rename(columns={0: 'Uncleaned_tokens', 1: 'Occurance Count'})
```

```python
sorted_token_count_plot = sorted_token_count.head(10)
plt.figure(figsize=[10,10])
sns.barplot(x="Uncleaned_tokens",y='Occurance Count',data=sorted_token_count_plot)
plt.title("Count of Ucleaned tokens")
```

**4. Cleaning Regulation Description Data (convert to lower case, remove stopwords, punctuation, numbers, etc)**

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
cleaned_token_count={}
cleaned_words=[]
for text in reg_discrip:
    words = nltk.tokenize.word_tokenize(text)
    stopwords = set(nltk.corpus.stopwords.words('english'))
    words = [word for word in words if len(word) > 1]
    words = [word for word in words if not word.isnumeric()]
    words = [word for word in words if word.isalpha()]
    words = [word.lower() for word in words]
    words = [word for word in words if word not in stopwords]
    
    for word in words:
        cleaned_words.append(word)
        if word in cleaned_token_count:
            cleaned_token_count[word]+=1
        else:
            cleaned_token_count[word]=1
```

```python
cleaned_token_count
```

**5. Finding top-10 tokens again (Cleaned)**

```python
import operator 
sorted_ctoken_count = sorted(cleaned_token_count.items(),key=operator.itemgetter(1),reverse=True)

sorted_ctoken_count = pd.DataFrame(sorted_ctoken_count)
sorted_ctoken_count = sorted_ctoken_count.rename(columns={0: 'Cleaned_tokens', 1: 'Occurance Count'})
sorted_ctoken_count.head(10)
```

```python
sorted_ctoken_count_plot = sorted_ctoken_count.head(10)
plt.figure(figsize=[10,10])
sns.barplot(y="Cleaned_tokens",x='Occurance Count',data=sorted_ctoken_count_plot,palette="Paired")
plt.title("Count of cleaned tokens")
```

**6. Finding top-10 tokens after applying Porter stemming to the tokens obtained in step 4.**

```python
ps_token_count={}
porter = nltk.PorterStemmer()
for word in cleaned_words:
    ps_word = porter.stem(word)
    
    if ps_word in ps_token_count:
        ps_token_count[ps_word]+=1
    else:
        ps_token_count[ps_word]=1  
```

```python
import operator 
sorted_pstoken_count = sorted(ps_token_count.items(),key=operator.itemgetter(1),reverse=True)

sorted_pstoken_count = pd.DataFrame(sorted_pstoken_count)
sorted_pstoken_count = sorted_pstoken_count.rename(columns={0: 'PS_tokens', 1: 'Occurance Count'})
sorted_pstoken_count.head(10)
```

```python
sorted_pstoken_count_plot = sorted_pstoken_count.head(10)
plt.figure(figsize=[10,10])
sns.barplot(y="PS_tokens",x='Occurance Count',data=sorted_pstoken_count_plot,palette="husl")
plt.title("Count of cleaned tokens after applying Porter Stemmming")
```

**7. Finding top-10 tokens after applying Lancaster stemming to the tokens obtained in step 4.**

```python
ls_token_count={}
lancaster = nltk.LancasterStemmer()
for word in cleaned_words:
    ls_word = lancaster.stem(word)
    
    if ls_word in ls_token_count:
        ls_token_count[ls_word]+=1
    else:
        ls_token_count[ls_word]=1
        
import operator 
sorted_lstoken_count = sorted(ls_token_count.items(),key=operator.itemgetter(1),reverse=True)

sorted_lstoken_count = pd.DataFrame(sorted_lstoken_count)
sorted_lstoken_count = sorted_lstoken_count.rename(columns={0: 'LS_tokens', 1: 'Occurance Count'})
sorted_lstoken_count.head(10)
```

```python
sorted_lstoken_count_plot = sorted_lstoken_count.head(10)
plt.figure(figsize=[10,10])
sns.barplot(y="LS_tokens",x='Occurance Count',data=sorted_lstoken_count_plot,palette="rocket")
plt.title("Count of cleaned tokens after applying Lancaster Stemmming")
```

**8. Find top-10 tokens after applying lemmatization to the tokens obtained in step 4.**

```python
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet 
lem_token_count={}
#wnl = nltk.WordNetLemmatizer()
lemmatizer = WordNetLemmatizer()
for word in cleaned_words:
    lem_word = lemmatizer.lemmatize(word,wordnet.VERB)
    
    if lem_word in lem_token_count:
        lem_token_count[lem_word]+=1
    else:
        lem_token_count[lem_word]=1
        
import operator 
sorted_lemtoken_count = sorted(lem_token_count.items(),key=operator.itemgetter(1),reverse=True)

sorted_lemtoken_count = pd.DataFrame(sorted_lemtoken_count)
sorted_lemtoken_count = sorted_lemtoken_count.rename(columns={0: 'LEM_tokens', 1: 'Occurance Count'})
sorted_lemtoken_count.head(10)
```

```python
sorted_lemtoken_count_plot = sorted_lemtoken_count.head(10)
plt.figure(figsize=[10,10])
sns.barplot(y="LEM_tokens",x='Occurance Count',data=sorted_lemtoken_count_plot,palette="crest")
plt.title("Count of cleaned tokens after applying Lemmatization")
```

**9. Comparing top-10 tokens obtained in step 3, 5, 6, 7, 8.**

In step 3, when we find out the top 10 tokens for the Regulation Description without cleaning the data and removing the noise, we notice that most of these tokens correspond to punctuations, stopwords and very less tokens give us an idea of what leads to the failed food inspections. (Reference to the table and graph in step 3)

In step 5, after cleaning the data and removing noise (punctuations, numbers, stopwords etc), we can see that (Refer Graph in step 5) we have a better picture of what the failed food inspections in Chicago were about. These tokens mostly relate to food, maintainence, cleanliness, equipments, construction and hygiene in general. 

In step 6, when we apply porter stemming to get the root word for cleaned tokens, we see a sudden jump in the tokens related to cleanliness("clean") when compared to step 5. We can infer from this that most failed inspections have descriptions related to cleanliness followed by food, maintainence, equip, construct, instal, surfac. We can also see that some of the tokens have lost their semantic meaning when porter stemming is applied. 

In step 7, when we apply lancaster stemming to get the root word for cleaned tokens, we see minor changes when compared to step 6. While intuitively we know the meaning of the root words when comparing it to porter stemming, the application of lancaster stemming has resulted in the loss of semantic meaning for most of these words/tokens. 

In Step 8, when we apply lemmatization on the cleaned tokens, we see that the results are somewhat similar to what we saw in step 6(porter stemming). However, semantic meaning is preserved and the results are more accurate. The tokens that lead to failed food inspections in Chicago mostly relate to cleanliness("clean"), maintainence("maintain"), food equipment, construction("construct") which makes sense. 


Finally, we see that lemmatization is a good technique when we want to understand the context from the tokens. Otherwise, stemming is a good technique because it is fast when compared to lemmatization.
